```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the Excel file (update path as needed)
file_path = '/content/drive/MyDrive/photometory data/Photometry of SN 2022eyw in UBV R bands (Vega system)..xlsx'
df = pd.read_excel(file_path)

# Strip spaces from column names
df.columns = df.columns.str.strip()

# Extract the JD column (Julian Date)
jd = df['JD†']

# Function to parse magnitude and magnitude error from strings like '16.81 ± 0.09' ignoring missing values
def extract_mag_err_ignore_missing(column):
    mags = []
    errs = []
    for val in column:
        if isinstance(val, str) and '±' in val:
            mag, err = val.split('±')
            try:
                mg = float(mag.strip())
                er = float(err.strip())
                mags.append(mg)
                errs.append(er)
            except ValueError:
                mags.append(np.nan)
                errs.append(np.nan)
        else:
            # For '', '-', None or other missing entry, treat as NaN
            mags.append(np.nan)
            errs.append(np.nan)
    return np.array(mags), np.array(errs)

# Define filters and their colors (V is violet)
filters = ['U', 'B', 'V', 'R']
colors = {'U': 'blue', 'B': 'cyan', 'V': 'violet', 'R': 'red'}

# Extract mags and errors for each filter ignoring missing values
filter_data = {f: extract_mag_err_ignore_missing(df[f]) for f in filters}

# Plot each filter magnitude vs JD with error bars individually with color
for f in filters:
    mags, errs = filter_data[f]
    mask = ~np.isnan(mags)
    plt.figure()
    plt.errorbar(jd[mask], mags[mask], yerr=errs[mask], fmt='o', color=colors[f], label=f'Filter {f}')
    plt.gca().invert_yaxis()  # Magnitude scale is reversed in astronomy
    plt.xlabel('Julian Date (JD)')
    plt.ylabel('Magnitude')
    plt.title(f'Light curve in {f}-band')
    plt.legend()
    plt.show()

# Plot all filters on one plot with respective colors
plt.figure()
for f in filters:
    mags, errs = filter_data[f]
    mask = ~np.isnan(mags)
    plt.errorbar(jd[mask], mags[mask], yerr=errs[mask], fmt='o', color=colors[f], label=f'Filter {f}')
plt.gca().invert_yaxis()
plt.xlabel('Julian Date (JD)')
plt.ylabel('Magnitude')
plt.title('Multi-filter Light Curves')
plt.legend()
plt.show()
```

```python

# Program to find jd_max, m_max, M_max, delta_m15


import pandas as pd
from scipy.interpolate import UnivariateSpline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from astropy.stats import sigma_clipped_stats
import extinction


file_path = '/content/drive/MyDrive/curve ftting/SN2022eyw_B.dat'
df = pd.read_csv(file_path, delim_whitespace=True)  # or sep='\t' if tab separated

# then access columns as usual, e.g.:
jd = df['JD']
mag = df['MAG']  # or your filter column name
mag_err=df['EMAG']


# --- Fit initial spline ---
spline = UnivariateSpline(jd, mag, k=4, s=0.1)
fine_jd = np.linspace(jd.min(), jd.max(), 1000)
fine_mag = spline(fine_jd)

# Find jd_max and mag_max from the spline
jd_max = fine_jd[np.argmin(fine_mag)]
mag_max = np.min(fine_mag)

delta_m15 = spline(jd_max + 15) - mag_max

# --- Monte Carlo Simulation ---
n_iter = 10000
jd_max_sim = []
mag_max_sim = []
delta_m15_sim = []

for i in range(n_iter):
    mag_sim = mag + np.random.normal(0, mag_err)
    spline_sim = UnivariateSpline(jd, mag_sim, k=4, s=0.1)
    fine_mag_sim = spline_sim(fine_jd)

    jd_max_sim.append(fine_jd[np.argmin(fine_mag_sim)])
    mag_max_sim.append(np.min(fine_mag_sim))

    # Δm15: magnitude 15 days after jd_max
    jd_plus_15 = fine_jd[np.argmin(fine_mag_sim)] + 15
    mag_plus_15 = spline_sim(jd_plus_15)
    delta_m15_sim.append(mag_plus_15 - np.min(fine_mag_sim))

# Convert to numpy arrays
jd_max_sim = np.array(jd_max_sim)
mag_max_sim = np.array(mag_max_sim)
delta_m15_sim = np.array(delta_m15_sim)

# --- Sigma-Clipped Stats ---

jd_max_mean, _, jd_max_std = sigma_clipped_stats(jd_max_sim)
mag_max_mean, _, mag_max_std = sigma_clipped_stats(mag_max_sim)
delta_m15_mean, _, delta_m15_std = sigma_clipped_stats(delta_m15_sim)

#jd_max_mean = np.mean(jd_max_sim)
#jd_max_std = np.std(jd_max_sim)

#mag_max_mean = np.mean(mag_max_sim)
#mag_max_std = np.std(mag_max_sim)

#delta_m15_mean = np.mean(delta_m15_sim)
#delta_m15_std = np.std(delta_m15_sim)

# --- Plot ---
plt.figure(figsize=(10,6))
plt.errorbar(jd, mag, yerr=mag_err, fmt='o', label='Observed Data', capsize=3)
plt.plot(fine_jd, fine_mag, '-', label='Spline Fit')
plt.axvline(jd_max, color='r', linestyle='--', label=f'jd_max = {jd_max:.2f}')
plt.axhline(mag_max, color='g', linestyle='--', label=f'mag_max = {mag_max:.2f}')

# Mark jd_max + 15
mag_15 = spline(jd_max + 15)
plt.plot(jd_max + 15, mag_15, 's', color='purple', label='mag @ jd_max + 15')

plt.gca().invert_yaxis()
plt.xlabel('Julian Date')
plt.ylabel('Apparent Magnitude (B band)')
plt.title('Supernova B-band Light Curve and Spline Fit')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Calculate Absolute Magnitude

E_BV = 0.072
mod = 33.13
mod_err = 0.38

Av = 3.1*E_BV
wave = np.array([4363.2])
ext = extinction.fm07(wave, Av)[0]

abs_mag_max = mag_max_mean - mod - ext
abs_mag_max_err = np.sqrt((mag_max_std)**2 + (mod_err)**2)


# --- Output Results ---
print(jd_max_mean)
print(f"JD of Maximum Brightness (jd_max): {jd_max:.3f} ± {jd_max_std:.3f}")
print(f"Maximum B-band Apparent Magnitude (mag_max): {mag_max_mean:.3f} ± {mag_max_std:.3f}")
print(f"Maximum B-band Absolute Magnitude (mag_max): {abs_mag_max:.3f} ± {abs_mag_max_err:.3f}")
print(f"Δm15 (B band): {delta_m15_mean:.3f} ± {delta_m15_std:.3f}")
```

```python

# Program to find jd_max, m_max, M_max, delta_m15


import pandas as pd
from scipy.interpolate import UnivariateSpline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from astropy.stats import sigma_clipped_stats
import extinction


file_path = '/content/drive/MyDrive/curve ftting/SN2022eyw_r.dat '
df = pd.read_csv(file_path, delim_whitespace=True)  # or sep='\t' if tab separated

# then access columns as usual, e.g.:
jd = df['JD']
mag = df['MAG']  # or your filter column name
mag_err=df['EMAG']


# --- Fit initial spline ---
spline = UnivariateSpline(jd, mag, k=4, s=0.1)
fine_jd = np.linspace(jd.min(), jd.max(), 1000)
fine_mag = spline(fine_jd)

# Find jd_max and mag_max from the spline
jd_max = fine_jd[np.argmin(fine_mag)]
mag_max = np.min(fine_mag)

delta_m15 = spline(jd_max + 15) - mag_max

# --- Monte Carlo Simulation ---
n_iter = 10000
jd_max_sim = []
mag_max_sim = []
delta_m15_sim = []

for i in range(n_iter):
    mag_sim = mag + np.random.normal(0, mag_err)
    spline_sim = UnivariateSpline(jd, mag_sim, k=4, s=0.1)
    fine_mag_sim = spline_sim(fine_jd)

    jd_max_sim.append(fine_jd[np.argmin(fine_mag_sim)])
    mag_max_sim.append(np.min(fine_mag_sim))

    # Δm15: magnitude 15 days after jd_max
    jd_plus_15 = fine_jd[np.argmin(fine_mag_sim)] + 15
    mag_plus_15 = spline_sim(jd_plus_15)
    delta_m15_sim.append(mag_plus_15 - np.min(fine_mag_sim))

# Convert to numpy arrays
jd_max_sim = np.array(jd_max_sim)
mag_max_sim = np.array(mag_max_sim)
delta_m15_sim = np.array(delta_m15_sim)

# --- Sigma-Clipped Stats ---

jd_max_mean, _, jd_max_std = sigma_clipped_stats(jd_max_sim)
mag_max_mean, _, mag_max_std = sigma_clipped_stats(mag_max_sim)
delta_m15_mean, _, delta_m15_std = sigma_clipped_stats(delta_m15_sim)

#jd_max_mean = np.mean(jd_max_sim)
#jd_max_std = np.std(jd_max_sim)

#mag_max_mean = np.mean(mag_max_sim)
#mag_max_std = np.std(mag_max_sim)

#delta_m15_mean = np.mean(delta_m15_sim)
#delta_m15_std = np.std(delta_m15_sim)

# --- Plot ---
plt.figure(figsize=(10,6))
plt.errorbar(jd, mag, yerr=mag_err, fmt='o', label='Observed Data', capsize=3)
plt.plot(fine_jd, fine_mag, '-', label='Spline Fit')
plt.axvline(jd_max, color='r', linestyle='--', label=f'jd_max = {jd_max:.2f}')
plt.axhline(mag_max, color='g', linestyle='--', label=f'mag_max = {mag_max:.2f}')

# Mark jd_max + 15
mag_15 = spline(jd_max + 15)
plt.plot(jd_max + 15, mag_15, 's', color='purple', label='mag @ jd_max + 15')

plt.gca().invert_yaxis()
plt.xlabel('Julian Date')
plt.ylabel('Apparent Magnitude (r band)')
plt.title('Supernova r-band Light Curve and Spline Fit')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Calculate Absolute Magnitude

E_BV = 0.072
mod = 33.13
mod_err = 0.38

Av = 3.1*E_BV
wave = np.array([4363.2])
ext = extinction.fm07(wave, Av)[0]

abs_mag_max = mag_max_mean - mod - ext
abs_mag_max_err = np.sqrt((mag_max_std)**2 + (mod_err)**2)


# --- Output Results ---
print(jd_max_mean)
print(f"JD of Maximum Brightness (jd_max): {jd_max:.3f} ± {jd_max_std:.3f}")
print(f"Maximum r-band Apparent Magnitude (mag_max): {mag_max_mean:.3f} ± {mag_max_std:.3f}")
print(f"Maximum r-band Absolute Magnitude (mag_max): {abs_mag_max:.3f} ± {abs_mag_max_err:.3f}")
print(f"Δm15 (r band): {delta_m15_mean:.3f} ± {delta_m15_std:.3f}")

# Mark jd_max + 15
mag_15 = spline(jd_max + 15)
plt.plot(jd_max + 15, mag_15, 's', color='purple', label='mag @ jd_max + 15')

plt.gca().invert_yaxis()
plt.xlabel('Julian Date')
plt.ylabel('Apparent Magnitude (g-band)')
plt.title('Supernova g-band Light Curve and Spline Fit')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure()
plt.hist(mag_err, 50, density=True)
plt.show()
```
